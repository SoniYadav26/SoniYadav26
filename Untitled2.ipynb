Importing the libraries.

```python
import numpy as np
import pandas as pd          # to read the csv file into more structured form
import difflib             # to get the closest match of the value
from sklearn.feature_extraction.text import TfidfVectorizer   # it is used to convert textual data into numerical value
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and pre-processing.

```python
# load the data from csv file to pandas dataframe by using pd library.
movies_data = pd.read_csv('/content/movies.csv')
```

```python
# printing the first 5 rows of dataframe.
movies_data.head()
```

```python
# check how many rows and column in a dataframe.
movies_data.shape
```

```python
# Select the relevant features for recommendation.
choosed_features = ['genres','title','cast','director','keywords']
print(choosed_features)
```

```python
# here we replacing the null values with null string.
for feature in choosed_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# here we are going to combinig the all 5 selected features.
combined_features = movies_data['genres']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['keywords']
```

```python
print(combined_features)
```

```python
# use Tfidvectorizer AS vectorizer.
vectorizer = TfidfVectorizer()
```

```python
# converrting the textual data to featued data or numerical data.
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine_similarity.

```python
# getting the similarity confidence value using cosine simliarity.
similarity_value = cosine_similarity(feature_vectors)
```

```python
print(similarity_value)
```

```python
# now here we are taking the movie name from the user.
movie_name = input('Enter ur favourite movie name: ')
```

```python
# print all the movie name from a list.
list_of_titles = movies_data['title'].tolist()
print(list_of_titles)
```

```python
# find the closest match of the movie name given by the user.
toget_closest_match = difflib.get_close_matches(movie_name,list_of_titles)
```

```python
print(toget_closest_match)
```

```python
close_match = toget_closest_match[0]
print(close_match)
```

```python
# finding the index of the movie with title.
index_of_movie = movies_data[movies_data.title == close_match].index[0]
```

```python
print(index_of_movie)
```

```python
# getting list of similar movies.

similarity_score= list(enumerate(similarity_value[index_of_movie]))
```

```python
print(similarity_score)
```

```python
# sorting the movies based on their similarity score.
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index.
print("Movies recommended for you: \n")

i=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index = movies_data[movies_data.index==index]['title']
  if(i<20):
    print(i,'.', title_from_index)
    i+=1
```

Movie recommended System.

```python
movie_name = input('Enter ur favourite movie name: ')

list_of_titles = movies_data['title'].tolist()

toget_closest_match = difflib.get_close_matches(movie_name,list_of_titles)

close_match = toget_closest_match[0]

index_of_movie = movies_data[movies_data.title == close_match].index[0]


similarity_score= list(enumerate(similarity_value[index_of_movie]))

sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse=True)

print("Movies recommended for you: \n")

i=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index = movies_data[movies_data.index==index]['title']
  if(i<20):
    print(i,'.', title_from_index)
    i+=1
```
